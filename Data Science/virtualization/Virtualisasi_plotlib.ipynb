```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,10]
y = [100,200,400,350,450,800]
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y, color='red',marker='o',markersize=10,linestyle='--');
plt.title("Grafik Coba")
plt.xlabel("Waktu")
plt.ylabel("Percepatan")
plt.xlim(0,10)
plt.ylim(0,1000)
plt.show()
```
